```python
from icrawler.builtin import BingImageCrawler
```

```python
bing_crawler = BingImageCrawler(
    downloader_threads=4,
    storage={'root_dir': './cnn_sample/cnn_sample/crawling'} # value에 경로 입력
)
# filter
```

```python
bing_crawler.crawl(keyword='Unripe tomatoes', filters=None, offset=0, max_num=1000) # keyword에 검색어, max_num = 최대 파일수
```
